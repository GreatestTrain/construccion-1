```python
import camelot
import pandas as pd
import re
```

```python
pdf_path = "file.pdf"
tables = camelot.read_pdf(pdf_path, pages="15-18")
```

```python
expr1 = re.compile(r'(?P<ID>[\w\d\.\*]+) \| (?P<Descripción>.+)')
expr2 = re.compile(r'.+ (?P<Rendimiento>[\d\.,]+)')
expr3 = re.compile(r'(?P<Unidad>.+)')
buff = []
for table in tables:
    val = expr1.search(table.df.loc[0,0]).groupdict()
    rend = expr2.search(table.df.loc[1,0]).groupdict()
    unit = expr3.search(table.df.iloc[1,2])
    unit = unit.groupdict() if unit else {}
    buff.append(val | rend | unit)
buff = pd.DataFrame(buff).fillna("PZA")
buff
```

```python
buff.to_csv("rendimientos.csv")
```

```python
df = []
for table in map(lambda x: x.df, tables):
    table["Partida"] = expr1.search(table.iloc[0,0])["Descripción"]
    table = table.drop([0,1], axis=0)
    table = table.rename(columns={
        0: "ID",
        1: "Descripción",
        2: "Unidad",
        3: "Cantidad",
    })
    table = table[["ID", "Descripción", "Partida", "Unidad", "Cantidad"]]
    df.append(table)
df = pd.concat(df)
df = pd.merge(left=df, right=buff[["ID", "Descripción"]], left_on="Partida", right_on="Descripción", how="left")\
    .rename(columns={
        "ID_x": "ID_Elemento",
        "ID_y": "ID_Partida",
        "Descripción_x": "Descripción",
    })[["ID_Partida", "ID_Elemento", "Descripción", "Unidad", "Cantidad"]]
```

```python
mask = df["ID_Elemento"].str.contains("\n")
if mask.any():
    cantidad = df["Unidad"].copy(deep=True)
    unidad = df["Descripción"].copy(deep=True)
    df.loc[mask, "Cantidad"] = cantidad
    df.loc[mask, "Unidad"] = unidad
    df.loc[mask, ["ID_Elemento", "Descripción"]] = df.loc[mask, "ID_Elemento"].str.split("\n", expand=True).values
    df.loc[mask]
```

```python
df.to_csv("partidas-elementos.csv")
df
```
